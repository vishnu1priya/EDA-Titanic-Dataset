### Fill the missing values by

   - Mean : When the data is having normal distribution (if mean = median, fillby mean)
   - Median: when the data is not having normal distribution (if mean != median, fill by median)

```python
df.Age.describe()
```

```python
# let's check the normal distribution

sns.distplot(df.Age)
plt.axvline(df.Age.mean(), color ='red')
plt.axvline(df.Age.median(), color = 'black')
plt.show()
```

```python
sns.boxplot(df.Age)
plt.axhline(df.Age.mean(), color = 'red')
plt.axhline(df.Age.median(), color = 'black')
plt.show()
```

#### Here little bit skeweness is present and the data isn't having the normal distribution
#### Hence we choose median to replace the values

```python
df.Age.fillna(df.Age.median(), inplace = True)
```

```python
df.isnull().sum().sort_values(ascending = True)
```

#### Feature Engineering

Create a new feature named __FamilySize__

```python
df['FamilySize'] = df.SibSp + df.Parch + 1
df.head(3)
```

##### Segment Sex Column as per Age.

___If age<15 =>child, else Sex___

```python
df.sample(6)
```

```python
df['Gender_Class'] = df.apply(lambda x: 'child' if x['Age']<15 else x['Sex'], axis = 1)
df.sample(6)
```
